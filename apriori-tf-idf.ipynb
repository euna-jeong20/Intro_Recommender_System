```python
import mlxtend
import numpy as np
import pandas as pd
```

```python
data = np.array([
                 ['우유', '기저귀', '쥬스'],        #0번째 거래
                 ['양상추', '기저귀', '맥주'],      #1번째 거래
                 ['우유', '양상추', '기저귀', '맥주'],      #2번째 거래
                 ['양상추', '맥주']                 #3번째 거래
])
```

# Apriori 알고리즘

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit_transform(data)
te_ary
```

```python
te.columns_
```

```python
df = pd.DataFrame(te_ary, columns= te.columns_)
df
```

```python
%%time
from mlxtend.frequent_patterns import apriori

apriori(df, min_support=0.5, use_colnames=True)
```

```python
%%time
from mlxtend.frequent_patterns import fpgrowth


fpgrowth(df, min_support=0.5, use_colnames=True)
```

```python
from mlxtend.frequent_patterns import association_rules 

association = apriori(df, min_support=0.1, use_colnames=True)
association_rules(association, metric = 'confidence', min_threshold=0.5, support_only=False)
```

```python
association = fpgrowth(df, min_support=0.1, use_colnames=True)
association_rules(association, metric = 'confidence', min_threshold=0.5, support_only=False)
```

# 원 핫 인코딩

```python
docs = [
    '먹고 싶은 사과', #문서0
    '먹고 싶은 바나나',  #문서1
    '길고 노란 바나나 바나나',  #문서2
    '저는 과일이 좋아요'   #문서3
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer()   #CountVectorizer 객체 생성
```

```python
countvect = vect.fit_transform(docs)
countvect
```

```python
# sparse matrix -> numpy
countvect.toarray()
```

```python
vect.vocabulary_
```

```python
#sorted로 value값을 기준으로 순서 정렬 
sorted(vect.vocabulary_)
```

```python
countvect_df = pd.DataFrame(countvect.toarray(), columns = sorted(vect.vocabulary_))  #넘파이 배열로 넘겨주기
countvect_df.index = ['문서0', '문서1', '문서2', '문서3']
countvect_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity(countvect_df, countvect_df)
```

# TF-IDF 알고리즘


* TF(d,t) : 특정 문서 d에서의 특정 단어 t의 등장 횟수
* DF(t) : 특정 단어 t가 등장한 문서의 수
* IDF(d,t) : DF에 반비례하는 수
* TF(d,t) * IDF(d,t)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vect = TfidfVectorizer()
tfvect = vect.fit_transform(docs)
```

```python
tfidv_df = pd.DataFrame(tfvect.toarray(), columns = sorted(vect.vocabulary_))
tfidv_df.index = ['문서0', '문서1', '문서2', '문서3']
tfidv_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity(tfidv_df, tfidv_df)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vect = TfidfVectorizer(max_features=5)
tfvect = vect.fit_transform(docs)
```

```python
tfidv_df = pd.DataFrame(tfvect.toarray(), index = ['문서0', '문서1', '문서2', '문서3'], columns = sorted(vect.vocabulary_))
tfidv_df
```
